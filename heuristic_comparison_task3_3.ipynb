### Heuristic Comparison: Task 2 vs Task 3.3

```python

def heuristic_euclidean(node, goal, points):
    x1, y1 = points[node]
    x2, y2 = points[goal]
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5 * 1000
```

```python

def heuristic_env_aware(node, goal, points, current, neighbor, constraints):
    base_dist = ((points[goal][0] - points[node][0]) ** 2 + (points[goal][1] - points[node][1]) ** 2) ** 0.5 * 1000

    edge = (current, neighbor) if (current, neighbor) in constraints else (neighbor, current)
    terrain = constraints.get(edge, None)

    penalty = 0
    if terrain == 'stairs':
        penalty = 5
    elif terrain == 'slope':
        penalty = 2
    elif terrain == 'lift':
        penalty = 4
    elif terrain == 'kerb_ramp':
        penalty = 3
    elif terrain == 'obstacle':
        penalty = 6  # severe penalty to avoid

    return base_dist * penalty
```
