### Iterators

Python iterators are a fundamental concept in the Python programming language. Iterators are used to iterate over a sequence of values, such as a list, tuple, or string, one item at a time. Iterators are an efficient way to access and process large amounts of data without loading the entire dataset into memory at once. This is a very useful feature for AI programming.



Python provides several built-in functions and data types that support iteration, including range(), enumerate(), zip(), and iterable data types like lists, tuples, and strings. Understanding Python iterators is essential for writing efficient and readable code, especially when working with large datasets or performing complex data manipulation tasks.

````python
print("**iterable list**")
fruits_list = ['apple', 'banana', 'cherry']
for fruit in fruits_list:
    print(fruit)

print("")
print("**iterable string**")
a_string = 'hello'
for letter in a_string:
    print(letter)


print("")
print("**range()**")
for i in range(5):
    print(i)

print("")
print("**enumerate()**")
for i, fruit in enumerate(fruits_list):
    print(i, fruit)

print("")
print("**zip()**")
prices = [0.99, 0.50, 1.99]
for fruit, price in zip(fruits_list, prices):
    print(fruit, price)
````

In Python, iterators are implemented using the Iterator protocol, which requires two methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, while the __next__() method returns the next value in the sequence. If there are no more values in the sequence, the __next__() method raises the StopIteration exception.

````python
class MyIterator:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

it = MyIterator(1, 5)
for i in it:
    print(i)
````

You may wonder why we called **lists** an "iterable" data type. Because an iterable, although can be iterated over using a loop, it is not automatically an iterator, meaning that we can use `next()` on it.

The `iter()` function is a built-in function that creates an iterator from an iterable object, such as a list, tuple, or string.

Try run `iter()` and `next()` on fruits_list. What happens?

````python
iter(fruits_list)
````

````python
next(fruits_list)
````

The list itself is not an iterator, so calling `next()` directly on fruits_list raises a `TypeError` . To make it work, you need to convert the list into an iterator first using `iter()`. Here is sample code.

````python
fruits_iterator = iter(fruits_list)  # Convert list to an iterator

print(next(fruits_iterator))  # Expected output: apple
print(next(fruits_iterator))  # Expected output: banana
print(next(fruits_iterator))  # Expected output: cherry
````

### Generator expressions and list comprehensions

List comprehensions and generator expressions are two related features in Python that allow you to create iterable objects in a concise and efficient way.

List comprehensions are a way to create a new list by applying an expression to each element of an existing list or other iterable object. The basic syntax of a list comprehension is as follows:

```
new_list = [expression for item in iterable if condition]
```

Here is an example for outputting the squares of the numbers 1 through 4 (Also think about the range() and if condition is essential or not)

````python
squares = [i**2 for i in range(1, 11) if i < 5]
print(squares)  # prints [1, 4, 9, 16]
````

Generator expressions are similar to list comprehensions, but instead of creating a new list, they create a generator object. This is very useful when you have a large dataset and only limited memory.

Note the syntax difference.
```
new_generator = (expression for item in iterable if condition)
```

````python
squares_generator = (i**2 for i in range(1, 11) if i < 5)
for square in squares_generator:
    print(square)
````

### Generator functions

Study this example.

```
class MyIterator:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

it = MyIterator(1, 5)
```

It can be replaced with a generator function, which has a more concise syntax.

````python
def my_generator(start, end):
  current = start
  while current <= end:
    yield current
    current += 1

it = my_generator(1, 5)

for i in it:
    print(i)
````

### Map, filter and reduce

map(), filter(), and reduce() are built-in functions in Python that are commonly used to perform operations on collections of data.

map(): This function applies a given function to each element of an iterable (such as a list or tuple) and returns a new iterable with the results.


````python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
````



filter(): This function applies a given function to each element of an iterable and returns a new iterable with only the elements that pass the function's test (i.e., return True).


````python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]
````


reduce(): This function applies a given function to the first two elements of an iterable, then applies the same function to the result and the next element, and so on, until all elements have been processed and a single result is returned. The syntax for reduce() is slightly different from the others, as it is not a built-in function in Python 3 and must be imported from the functools module.

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x*y, numbers)
print(product)  # Output: 120
````
